### Random Walk

The basis of Brownian Motion is a Random Walk. A random walk in 1-D is defined as follows:

* A simple random walk in 1-D is when a step forward ($+d$ distance) has probability $p$ and a step back ($-d$ distance) has probability $q = 1 - p$.
* This can be extended to multiple dimensions quite easily using the same logic as in the 1-D case.

Random walks are an example of Markov processes, in which future behaviour is independent of past history.

```python
import random
import numpy as np
import plotly.express as px

def simulate_1d_rw(nsteps=1000, p=0.5, stepsize=1):
    steps = [ 1*stepsize if random.random() < p else -1*stepsize for i in range(nsteps) ]
    y = np.cumsum(steps)
    x = list(range(len(y)))

    return x, list(y)

simulation_data = {}
nsims = 5
for i in range(nsims):
    x, y = simulate_1d_rw()
    simulation_data['x'] = x
    simulation_data['y{col}'.format(col=i)] = y

ycols = [ 'y{col}'.format(col=i) for i in range(nsims) ]
fig = px.line(simulation_data, x='x', y=ycols)
fig.show()
```

```python
import random
import numpy as np
import plotly.express as px

def simulate_2d(nsteps=10000, stepsize=1):

    deltas = [ (0,-1*stepsize), (-1*stepsize,0), (0,1*stepsize), (1*stepsize,0) ]

    steps = [ list(random.choice(deltas)) for i in range(nsteps) ]
    steps = np.array(steps)
    steps = np.cumsum(steps,axis=0)
    y = list(steps[:,1])
    x = list(steps[:,0])

    return x, y

x, y = simulate_2d()
fig = px.line({ 'x' : x, 'y' : y }, x='x', y='y')
fig.show()
```

```python
import random
import numpy as np
import plotly.express as px

def simulate_3d_rw(nsteps=10000, stepsize=1):

    deltas = [ (0,0,-1*stepsize), (0,-1*stepsize,0), (-1*stepsize,0,0),\
               (0,0,1*stepsize), (0,1*stepsize,0), (1*stepsize,0,0) ]

    steps = [ list(random.choice(deltas)) for i in range(nsteps) ]
    steps = np.array(steps)
    steps = np.cumsum(steps,axis=0)
    z = list(steps[:,2])
    y = list(steps[:,1])
    x = list(steps[:,0])

    return x, y, z

x, y, z = simulate_3d_rw()
fig = px.line_3d({ 'x' : x, 'y' : y, 'z' : z }, x='x', y='y', z='z')
fig.show()
```

Brownian motion can be derived as a limit of a random walk as the step size decreases to zero and time becomes continuous.

```python
import random
import numpy as np
import plotly.express as px

def simulate_1d_bm(nsteps=1000, t=0.01):
    steps = [ np.random.randn()*np.sqrt(t) for i in range(nsteps) ]
    y = np.cumsum(steps)
    x = [ t*i for i in range(nsteps) ]
    return x, y

nsims = 5
simulation_data = {}
for i in range(nsims):
    x, y = simulate_1d_bm()
    simulation_data['y{col}'.format(col=i)] = y
    simulation_data['x'] = x

ycols = [ 'y{col}'.format(col=i) for i in range(nsims) ]
fig = px.line(simulation_data, x='x', y=ycols)
fig.show()
```

```python
import random
import numpy as np
import plotly.express as px

def simulate_2d_bm(nsteps=1000, t=0.01):
    x = np.cumsum([ np.random.randn()*np.sqrt(t) for i in range(nsteps) ])
    y = np.cumsum([ np.random.randn()*np.sqrt(t) for i in range(nsteps) ])
    return list(x), list(y)

x, y = simulate_2d_bm()
fig = px.line({ 'x' : x, 'y' : y }, x='x', y='y')
fig.show()
```

```python
import random
import numpy as np
import plotly.express as px

def simulate_3d_bm(nsteps=10000, t=0.01):
    x = np.cumsum([ np.random.randn()*np.sqrt(t) for i in range(nsteps) ])
    y = np.cumsum([ np.random.randn()*np.sqrt(t) for i in range(nsteps) ])
    z = np.cumsum([ np.random.randn()*np.sqrt(t) for i in range(nsteps) ])
    return list(x), list(y), list(z)

x, y, z = simulate_3d_bm()
fig = px.line_3d({ 'x' : x, 'y' : y, 'z' : z }, x='x', y='y', z='z')
fig.show()
```

### Brownian Motion with Drift

This is a variation of Brownian motion where there is a drift component along with the usual random component.

Along with the random noise in $X(t)$, there is the term $\mu t$, which represents the drift.

$$X(t) = \mu t + \sigma W(t)$$

```python
import random
import numpy as np
import plotly.express as px

def simulate_1d_bm_with_drift(nsteps=1000, t=0.01, mu=0.5):
    steps = [ mu*0.01 + np.random.randn()*np.sqrt(t) for i in range(nsteps) ]
    y = np.cumsum(steps)
    x = [ t*i for i in range(nsteps) ]
    return x, y

nsims = 5
simulation_data = {}
for i in range(nsims):
    x, y = simulate_1d_bm_with_drift()
    simulation_data['y{col}'.format(col=i)] = y
    simulation_data['x'] = x

ycols = [ 'y{col}'.format(col=i) for i in range(nsims) ]
fig = px.line(simulation_data, x='x', y=ycols)
fig.show()
```

### Geometric Brownian Motion

This process is often used to model financial stock prices, population growth, or other situations where measurements cannot be negative. In finance, it forms the basis for the Black–Scholes equation where the log returns of stock prices are modelled: the main interest is in the option pricing of derivatives.

It is defined by the following stochastic differential equation:

$$dG(t) = \mu G(t) dt + \sigma G(t) dW(t)$$

Alternatively, the explicit solution can be written as:

$$G(t) = G(0) \exp\left( \left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W(t) \right)$$

Note that $G(t)$ itself is not a standard Brownian motion, but its natural logarithm, $\log[G(t)]$, is a Brownian motion with drift.

```python
import random
import numpy as np
import plotly.express as px

def simulate_1d_gbm(nsteps=1000, t=1, mu=0.0001, sigma=0.02):
    steps = [ (mu - (sigma**2)/2) + np.random.randn()*sigma for i in range(nsteps) ]
    y = np.exp(np.cumsum(steps))
    x = [ t*i for i in range(nsteps) ]
    return x, y

nsims = 5
simulation_data = {}
for i in range(nsims):
    x, y = simulate_1d_gbm()
    simulation_data['y{col}'.format(col=i)] = y
    simulation_data['x'] = x

ycols = [ 'y{col}'.format(col=i) for i in range(nsims) ]
fig = px.line(simulation_data, x='x', y=ycols)
fig.show()
```

### Simulating Stock Price using Geometric Brownian Motion

As seen the above definition we can use actual stock price data to estimate μ & σ and use the parameters to simulate the stock price. In the following example we will look at the Dow Jones index between 2022–01–01 and 2022–12–31 i.e for the year of 2022. We will estimate the parameters of the Geometric Brownian Motion and use it to simulate the price of the index.

```python
import yfinance as yf
import numpy as np
import plotly.express as px
import random

def simulate_1d_gbm(nsteps=1000, t=1, mu=0.0001, sigma=0.02, start=1):
    # Generamos los pasos basados en la fórmula del Movimiento Browniano Geométrico
    steps = [ (mu - (sigma**2)/2) + np.random.randn()*sigma for i in range(nsteps) ]
    y = start*np.exp(np.cumsum(steps))
    x = [ t*i for i in range(nsteps) ]
    return x, y

# Descargamos los datos con yfinance
df = yf.download('^DJI', start='2022-01-01', end='2022-12-31')

# yfinance ya devuelve los datos en orden cronológico (oldest to newest),
# extraemos los precios de cierre directamente a un array 1D
prices = df['Close'].to_numpy().flatten()

logprices = np.log(prices)
logreturns = logprices[1:] - logprices[:-1]

# Estimamos los parámetros del drift (mu) y la volatilidad (sigma) a partir de los datos reales
mu = np.mean(logreturns)
sigma = np.std(logreturns)
nsteps = logprices.shape[0]

# Simulamos nuestro modelo
x, y = simulate_1d_gbm(nsteps=nsteps, mu=mu, sigma=sigma, start=prices[0])

# Preparamos los datos para la gráfica
data = {}
data['x'] = x
data['Simulation'] = y
data['DowJonesIndex'] = prices

# Dibujamos la serie real frente a nuestra simulación
fig = px.line(data, x='x', y=['DowJonesIndex', 'Simulation'],
              title='Dow Jones 2022 vs Geometric Brownian Motion Simulation')
fig.show()
```
